# Data Cleaning

#### We need complete trajectories, even for persons that have rows including unusable data (errors, outliers, etc.). To do this, we will detele the erroneous values, and then extrapolate them linearly. This is done because time series predictive machine learning models require time index frequency to be consistent (in our case, every second).

## Clean erroneous data from dataframe (negative y, excessively high x) - replace with NaN for future linear extrapolation

```python
# Replace negative y_world values with NaN:

df.loc[(df.y_world <= 0), "y_world"] = np.NaN

# Replace x_world values < -15m with NaN:

df.loc[(df.x_world <= -15), "x_world"] = np.NaN

# Replace x_world values > 15m with NaN:

df.loc[(df.x_world >= 15), "x_world"] = np.NaN
```

### Custom function for feature engineering
- drop duplicated timestamp rows  
- linearly interpolate missing x/y world values  
- fill in missing data for identified gender and age
- assign one single gender/age by person
- total time present in seconds per person_id
- cumulative time present in seconds per person_id
- x/y_world velocity
- (x,y) velocity
- distance to camera
- 5 step exponential moving average velocity for x, y, (x,y) 
- 10 step exponential moving average velocity for x, y, (x,y) 
- update presence in n label for last n rows

#### Add a presence label column to the dataframe for later use in goal 1 (person presence in n steps: 1/0)

```python
df["present_in_1"] = 1
df["present_in_2"] = 1
df["present_in_5"] = 1
```

```python
def add_features(d):

    # drop duplicated timestamp rows, reset index at fixed 1 second intervals
    d = d.drop_duplicates(subset="local_timestamp")
    d = d.set_index("local_timestamp").asfreq(freq="1 S", method="ffill")
    d = d.reset_index()

    # Making sure rows are sorted by time, so that total_time calculation makes sense
    d = d.sort_values(by="local_timestamp", ascending=True)

    # linearly interpolate NaN y/x_world (created by cleaning the erroneous values)
    d["y_world"] = d["y_world"].interpolate(method="linear")
    d["x_world"] = d["x_world"].interpolate(method="linear")

    # If the last age/gender estimate is NaN, replace it with the previous estimate
    d["rolling_exp_gender"] = d["rolling_exp_gender"].ffill()
    d["rolling_exp_age"] = d["rolling_exp_age"].ffill()

    # Make sure the first gender/age estimate isn't NaN by filling it with the next estimate
    d["rolling_exp_gender"] = d["rolling_exp_gender"].bfill()
    d["rolling_exp_age"] = d["rolling_exp_age"].bfill()

    # As gender and age are rolling estimates, let's apply the last estimate (which is the most accurate) to all rows of each person_id,
    # so that each unique person has 1 gender and 1 age estimate (for modeling purposes)
    d["gender"] = d.iloc[-1]["rolling_exp_gender"]
    d["age"] = d.iloc[-1]["rolling_exp_age"]

    # Compute total time spent in zone by substracting start time from exit time
    total_time = max(d["local_timestamp"]) - min(d["local_timestamp"])
    d["total_time_spent_seconds"] = total_time.seconds  # Convert time delta to integer

    # Compute cumulative time spent in zone by substracting start time from current time
    cumulative_time = (d["local_timestamp"]) - min(d["local_timestamp"])
    d["cumulative_time"] = cumulative_time.dt.seconds  # Convert time delta to integer

    # compute delta x and y, velocity
    d["dx_world"] = d["x_world"].diff(-1)
    d["dy_world"] = d["y_world"].diff(-1)

    # compute time shift, time dela
    d["timestamp_shift"] = d["local_timestamp"].shift(-1)
    d["d_time"] = d["timestamp_shift"] - d["local_timestamp"]
    d["d_time"] = d["d_time"].apply(lambda x: x.seconds)

    # drop now useless x_world/y_world shift columns
    d = d.drop(["timestamp_shift"], axis=1)

    # compute x_velocity, y_velocity, velocity
    d["x_velocity"] = d["dx_world"] / d["d_time"]
    d["y_velocity"] = d["dy_world"] / d["d_time"]
    d["velocity"] = np.sqrt(d.dx_world ** 2 + d.dy_world ** 2) / d["d_time"]

    # here, camera is assumed to be at 0,0
    # compute distance to camera
    d["distance_to_camera"] = np.sqrt(d.x_world ** 2 + d.y_world ** 2)

    # compute exponential moving averages for 5 steps
    d["ema_5_x"] = d["x_velocity"].ewm(span=5).mean()
    d["ema_5_y"] = d["y_velocity"].ewm(span=5).mean()
    d["ema_5_velocity"] = d["velocity"].ewm(span=5).mean()

    # compute exponential moving averages for 10 steps
    d["ema_10_x"] = d["x_velocity"].ewm(span=10).mean()
    d["ema_10_y"] = d["y_velocity"].ewm(span=10).mean()
    d["ema_10_velocity"] = d["velocity"].ewm(span=10).mean()

    # Drop last row that result in NaN values for multiple features
    d.drop(d.tail(1).index, inplace=True)

    # Update person presence label to 0 for all rows where presence stops 1, 2, 5 steps later:
    d.tail(1)["present_in_1"] = 0
    d.tail(2)["present_in_2"] = 0
    d.tail(5)["present_in_5"] = 0

    return d
```

### Define custom function to return outlier range (in terms of total time spent)

```python
# define function to return outlier range
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    Q1, Q3 = np.percentile(datacolumn, [25, 75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range


# retun outlier lower and upper bands for verification
lowerbound, upperbound = outlier_treatment(new_data["total_time_spent_seconds"])

print(lowerbound)
print(upperbound)

if lowerbound < 3:
    lowerbound = float(
        3
    )  # based on our understanding of the real world situation, we chose 3sec and 2min boundaries
if upperbound > 120:
    upperbound = float(120)  # confirmed by company team

print(lowerbound)
print(upperbound)
```

```python
def plot_installation(df_inst, No_of_ppl, name):
    def plot_personpath(p_id):
        df_p = df_inst.loc[df["person_id"] == p_id]
        x_pos = df_p["x_world"]
        y_pos = df_p["y_world"]
        x_direct = df_p["dx_world"]
        y_direct = df_p["dy_world"]
        scl = np.mean((x_direct ** 2 + y_direct ** 2) ** 0.5)
        return (x_pos, y_pos, x_direct, y_direct, scl)

    persons = df["person_id"].unique()
    persons = persons[0::No_of_ppl]
    len(persons)
    from matplotlib.pyplot import cm

    n = len(persons)
    color = cm.jet(np.linspace(0, 1, n))
    fig, ax = plt.subplots(figsize=(10, 10))
    for p in range(len(persons)):
        x_pos, y_pos, x_direct, y_direct, scl = plot_personpath(persons[p])
        ax.quiver(
            x_pos,
            y_pos,
            x_direct,
            y_direct,
            scale_units="xy",
            angles="xy",
            scale=1,
            color=color[p],
            linewidth=0.5,
            headwidth=2,
            headlength=4,
        )
    ax.set_xlabel("x_world", fontsize=16)
    ax.set_ylabel("y_world", fontsize=16)
    ax.tick_params(axis="both", which="major", labelsize=14)
    ax.text(
        0,
        0,
        "A",
        fontsize=20,
        horizontalalignment="center",
        verticalalignment="center",
        bbox=dict(facecolor="white", edgecolor="none", alpha=1),
    )
    plt.xlim((-10, 10))
    plt.ylim((-2.5, 15))
    plt.savefig("all_path_" + name + ".png", dpi=100)
    plt1 = plt.show()
    plt.draw()
    return plt1


None
```

```python
plot_installation(df_inst1, 500, nameof(df_inst1))
```
