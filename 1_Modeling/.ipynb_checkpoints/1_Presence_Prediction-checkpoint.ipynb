### Downsampling / Rebalancing

### Define functions to rebalance data

```python
# to rebalance data for goal 1 we will keep 5 last rows untouched
# and from the head "keep_x_th_row" only


def rebalance_data(d, keep_x_th_row=5):

    d_head = d.head(-5)
    d_head = d_head[d_head.index % keep_x_th_row == 0]  # Keep every x_th row
    d_tail = d.tail(5)

    frames = [d_head, d_tail]
    d_balanced = pd.DataFrame()
    d_balanced = pd.concat(frames, sort=True)
    return d_balanced


# to rebalance data for goal 1 with this function we will keep "num" last rows untouched
# and from the head "num" of random rows only
def rebalance_data_50_50(d, num=1):
    if len(d) >= num * 2:

        d_head = d.head(-num)
        d_head = d_head.sample(n=num, random_state=42)  # Keeep only x random rows
        d_tail = d.tail(num)

        frames = [d_head, d_tail]
        d_balanced = pd.DataFrame()
        d_balanced = pd.concat(frames, sort=True)
        return d_balanced
```

```python
# applying rebalance_data_50_50 function to the grouped dataframe for 5, 2 and 1 sec window
# change num accordingly
df_balanced_50_50_5 = df_final.groupby(["poi_id", "person_id"], group_keys=False).apply(
    lambda x: rebalance_data_50_50(x, num=5)
)
```

## Create sub dataframes for each installation 
### Define function to split sub dataframes into X_train, y_train, X_test, y_test, X_valid, y_valid
X,y valid are 5 % of the dataset, they could be used for additional testing if we want gto give some person_id's a longer trajectory


```python
def define_target_per_poi(
    sec, df, columns=[],
):
#create sub dataframes per POI
    if columns == []:
        columns = [
            "poi_id",
            "person_id",
            "x_world",
            "y_world",
            "age",
            "gender",
            "looking_at_screen",
            "ema_5_x",
            "ema_5_y",
            "ema_5_velocity",
            "cumulative_time",
            "dx_world",
            "dy_world",
            "x_velocity",
            "y_velocity",
            "velocity",
            "distance_to_camera",
            "ema_10_x",
            "ema_10_y",
            "ema_10_velocity",
        ]

    columns.append("present_in_" + str(sec))
    df_model = df[columns]
    # print("The model target window is " + str(sec) + " seconds. \n")
    # print(df_model.info())

    df_inst_1 = df_model.loc[df["poi_id"] == 444]
    df_inst_450 = df_model.loc[df["poi_id"] == 450]
    df_inst_453 = df_model.loc[df["poi_id"] == 453]
    df_inst_456 = df_model.loc[df["poi_id"] == 456]
    df_inst_457 = df_model.loc[df["poi_id"] == 457]
    df_inst_466 = df_model.loc[df["poi_id"] == 466]
    df_inst_529 = df_model.loc[df["poi_id"] == 529]

    return (
        df_inst_1,
        df_inst_450,
        df_inst_453,
        df_inst_456,
        df_inst_457,
        df_inst_466,
        df_inst_529,
    )


def split_per_group(sec, d, df_full=df_final):
# split POI dataframes into train-test 
    df_90_inds, valid_inds = next(
        GroupShuffleSplit(test_size=0.05, n_splits=2, random_state=42).split(
            d, groups=d["person_id"]
        )
    )

    df_90 = d.iloc[df_90_inds]
    valid = d.iloc[valid_inds]
    new_valid = (
        df_final[df_final.person_id.isin(valid.person_id.unique())][
            [
                "poi_id",
                "person_id",
                "x_world",
                "y_world",
                "x_world",
                "y_world",
                "age",
                "gender",
                "looking_at_screen",
                "ema_5_x",
                "ema_5_y",
                "ema_5_velocity",
                "cumulative_time",
                "dx_world",
                "dy_world",
                "x_velocity",
                "y_velocity",
                "velocity",
                "distance_to_camera",
                "ema_10_x",
                "ema_10_y",
                "ema_10_velocity",
                "present_in_1",
                "present_in_2",
                "present_in_5",
            ]
        ]
        .groupby(["person_id"], group_keys=False)
        .tail(5)
    )

    train_inds, test_inds = next(
        GroupShuffleSplit(test_size=0.2, n_splits=2, random_state=42).split(
            df_90, groups=df_90["person_id"]
        )
    )

    train = df_90.iloc[train_inds]
    test = df_90.iloc[test_inds]

    y_train = train.iloc[:, -1]
    y_test = test.iloc[:, -1]
    y_valid = new_valid["present_in_" + str(sec)]

    # Drop present_in_sec, person_id and POI_id from our POI dataframes (not needed and leading to errors)
    X_train = train[train.columns[2:-1]]
    X_test = test[test.columns[2:-1]]
    X_valid = new_valid[new_valid.columns[2:-3]]

    return X_train, y_train, X_test, y_test, X_valid, y_valid
```

### Modeling Pipeline (transformers, preprocessor, model)

```python
# Define numerical features
numerical_features = [
    "x_world",
    "y_world",
    "age",
    "ema_5_x",
    "ema_5_y",
    "ema_5_velocity",
    "cumulative_time",
    "dx_world",
    "dy_world",
    "x_velocity",
    "y_velocity",
    "velocity",
    "distance_to_camera",
    "ema_10_x",
    "ema_10_y",
    "ema_10_velocity",
]

# Define categorical features
categorical_features = ["gender", "looking_at_screen"]

# Define transformers
numerical_transformer = Pipeline(steps=[("scaler", MinMaxScaler()),])

categorical_transformer = Pipeline(
    steps=[("onehot", OneHotEncoder(handle_unknown="ignore"))]
)

# Combine feature transformers
data_transformer = ColumnTransformer(
    transformers=[
        ("numerical", numerical_transformer, numerical_features),
        ("categorical", categorical_transformer, categorical_features),
    ],
    remainder="passthrough",
)

# Define preprocessing pipeline (data transformation & dimension reduction, not included for now)
preprocessor = Pipeline(steps=[("data_transformer", data_transformer),])

# Define model pipeline
rf_model = RandomForestClassifier()
rfc_pipe = Pipeline(steps=[("preprocessor", preprocessor), ("classifier", rf_model)])
```

## Model 2 - Logistic Regression

### Modeling Pipeline (transformers, preprocessor, model)

```python
# Define numerical features
numerical_features = [
    "x_world",
    "y_world",
    "age",
    #    "ema_5_x",
    #    "ema_5_y",
    #    "ema_5_velocity",
    "cumulative_time",
]

# Define categorical features
categorical_features = ["gender", "looking_at_screen"]

# Define transformers
numerical_transformer = Pipeline(
    steps=[("scaler", StandardScaler()),]
)  # possible to add an imputer as well

categorical_transformer = Pipeline(
    steps=[("onehot", OneHotEncoder(handle_unknown="ignore"))]
)  # possible to add label_encoder, and imputer

# Combine feature transformers
data_transformer = ColumnTransformer(
    transformers=[
        ("numerical", numerical_transformer, numerical_features),
        ("categorical", categorical_transformer, categorical_features),
    ],
    remainder="passthrough",
)  # features not listed in categorical/numerical lists will be ignored ("passthrough")

# Define preprocessing pipeline (data transformation & dimension reduction, not included for now)
preprocessor = Pipeline(steps=[("data_transformer", data_transformer),])

# Define model pipeline
log_model = LogisticRegression(
    class_weight="balanced", C=3593.813663804626, solver="liblinear", penalty="l2"
)
log_pipe = Pipeline(steps=[("preprocessor", preprocessor), ("classifier", log_model)])
```

```python
# Model fitting, predictions, and scoring

model_accuracy = {}


def run_model(model_name, X_train, y_train, X_test, y_test):

    log_fit = log_pipe.fit(X_train, y_train)
    predictions = log_pipe.predict(X_test)
    log_score = log_pipe.score(X_test, y_test)

    print(classification_report(y_test, predictions))
    print("----------------------------------------------------------")
    print(confusion_matrix(y_test, predictions))
    print("----------------------------------------------------------")
    skplt.metrics.plot_confusion_matrix(y_test, predictions, normalize=True)
    accuracy = accuracy_score(y_test, predictions)
    print("Accuracy: %.2f%%" % (accuracy * 100.0))
    print("----------------------------------------------------------")

    model_accuracy[model_name] = accuracy
```
